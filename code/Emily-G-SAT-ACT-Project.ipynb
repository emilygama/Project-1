## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*Source*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

While the initial substantial wave of test-optional universities was in response to the COVID-19 pandemic, the list continues to grow of schools that are opting for permanent policies (<a href="https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html" target="_blank">Source</a>, <a href="https://www.fairtest.org/record-1775-schools-are-testoptional-fall-2022" target="_blank">Source</a>). The push behind this is the belief that the tests neither accurately assess college-readiness, nor do they provide any diversity or equality among college applicants (<a href="https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html" target="_blank">Source</a>). There are myriad issues with the tests and their persistence. There is a financial disparity with the tests, favoring those that can afford prep courses. Further, research has shown that white students consistently outperform their minority peers, and it has been suggested these tests and similar teachings styles promote the school-to-prison pipeline (<a href="https://www.researchgate.net/publication/280232788_Race_Poverty_and_SAT_Scores_Modeling_the_Influences_of_Family_Income_on_Black_and_White_High_School_Students'_SAT_Performance" target="_blank">Source</a>, <a href="https://www.fairtest.org/how-testing-feeds-schooltoprison-pipeline" target="_blank">Source</a>). Not to mention, these tests leave a window for wealthy families to manipulate the system as evidenced by the now-infamous testing scandal that involved a few celebrity names (<a href="https://en.wikipedia.org/wiki/2019_college_admissions_bribery_scandal#:~:text=In%202019%2C%20a%20scandal%20arose,code%20named%20Operation%20Varsity%20Blues" target="_blank">Source</a>). 

While the lack of requirement to be admitted to college is a welcome and growing movement, there is less focus on the graduation requirement many states hold for their high school seniors. Twenty-six states maintain graduation requirements that mandate students take one of the tests (and cannot qualify with the other), choose between either, or as an alternative to state-created standardized tests (<a href="https://prepmaven.com/blog/test-prep/states-require-sat-act/" target="_blank">Source</a>). I argue that focus on removing the graduation requirement for high school seniors is important as well. With the introduction of the No Child Left Behind policies, standardized testing massively increased in schools, with districts searching for a fair way to manage and test student success (<a href="https://fairtest.org/how-standardized-testing-damages-education-pdf" target="_blank">Source</a>). However, standardized testing furthered by the No Child Left Behind policies and, specifically, the SAT/ACT requirements have consistently been a hindrance to student success than the opposite. Instead of pushing for a measure of learning and college-readiness, these tests and their requirements have created an environment in which students may feel immense pressure, stress, and feelings of worthlessness if they do not measure up to where they are supposed to be, often with little individualized help from administration (<a href="https://fairtest.org/how-standardized-testing-damages-education-pdf" target="_blank">Source</a>). 

While this argument applies to all standardized testing within the public education system, I want to focus specifically on the SAT and ACT for this analysis. One 30-year study that examined the same group of students from their middle school years into their thirties found that test anxiety and academic self-esteem concerns were widely pronounced and significant, mostly in their middle school years. They found that these concerns were brought on by many factors, largely including standardized testing and consequences of underperforming on a test (<a href="https://www.sciencedirect.com/science/article/abs/pii/S0165032717303683" target="_blank">Source</a>). Cassady and Johnson studied 168 high schoolers and found that 'high levels of cognitive test anxiety were associated with significantly lower Scholastic Aptitude Test scores' (<a href="https://www.sciencedirect.com/science/article/abs/pii/S0361476X0191094X" target="_blank">Source</a>). This project will aim to evaluate student performance in comparison to participation rates on the SAT and ACT test. Because I was unable to find open-source student mental health data to contribute to this project, I am proposing we apply for grant funding to do a large-scale study on numerous high school students from their preparation to graduate and into college to evaluate the impact academia makes upon mental health.

```python
temp_nums = [2,4,6,8,10,12,14,16]
def manual_mean(nums):
    return sum(nums)/len(nums)
manual_mean(temp_nums)
```

```python
def manual_stdev(nums):
    return (manual_mean([(num-manual_mean(nums))**2 for num in nums]))**0.5
manual_stdev(temp_nums)
```

```python
def clean_percents(string_percents):
    for x in string_percents:
        return float(string_percents.replace('%', ''))/100
clean_percents('47%')
```
